## Matrix Plot

### Heat map

```python
flights=sns.load_dataset('flights')
flights
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tipscorr=tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(), annot=True)
```

### Cluster Map

```python
sns.clustermap(tipscorr.corr(), annot=True, cmap='coolwarm')
```

### Pivot Table Heat Map

```python
flights= sns.load_dataset('flights')
flights
```

```python
pvtflight=flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
pvtflight
```

```python
sns.heatmap(pvtflight)
```
